# 1.LeetCode 225. Implement Stack using Queues

```python
class MyStack(object):

    def __init__(self):
        self.__queue1 = []
        self.__queue2 = []

    def push(self, x):
        self.__queue2.append(x)
        while (len(self.__queue1) != 0):
            self.__queue2.append(self.__queue1.pop(0))
        self.__queue1, self.__queue2 = self.__queue2, self.__queue1
        """
        :type x: int
        :rtype: None
        """

    def pop(self):
        return self.__queue1.pop(0)
        """
        :rtype: int
        """

    def top(self):
        return self.__queue1[0]
        """
        :rtype: int
        """

    def empty(self):
        if (len(self.__queue1) == 0):
            return True
        return False
        """
        :rtype: bool
        """
```
